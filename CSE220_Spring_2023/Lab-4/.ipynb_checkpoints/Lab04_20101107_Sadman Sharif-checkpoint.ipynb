```python
class Patient:
    
    def __init__(self,idt,name,age,bg):
        
        self.idt=idt
        self.name=name
        self.age=age
        self.bg=bg


class Doubly:
    
    def __init__(self, elem,nxt=None,prev=None):
        self.elem=elem
        self.next=nxt
        self.prev=prev


class WRM(Doubly):
    
    def __init__(self):
        
        self.head=Doubly(None)
        self.tail=Doubly(None)
        self.head.next=self.tail
        self.tail.prev=self.head

    def register_patient(self,idt,name,age,bg):
        
        new_patient = Patient(idt,name,age,bg)
        new_node = Doubly(new_patient)

        last_node=self.tail.prev
        last_node.next=new_node
        new_node.prev=last_node
        new_node.next=self.tail
        self.tail.prev=new_node

    def serve_patient(self):
        
        if self.head.next==self.tail and self.tail.prev==self.head:
            
            print("No patients are left in the waiting room.")
            return

        first_patient=self.head.next.elem
        first_node=self.head.next

        self.head.next=first_node.next
        first_node.next.prev=self.head

        print(first_patient.name,"is served")

    def cancel_all_appointments(self):
        
        if self.head.next==self.tail and self.tail.prev==self.head:
            
            print("No patients are left in the waiting room.")
            return

        self.head.next = self.tail
        self.tail.prev = self.head
        
        print("All appointments are cancelled.")

    def can_doctor_go_home(self):
        
        if self.head.next==self.tail and self.tail.prev==self.head:
            
            print("Yes, the doctor can go home.")
            
        else:
            print("No, patients are still waiting.")

    def show_all_patients(self):
        
        if self.head.next==self.tail and self.tail.prev==self.head:
            
            print("No patients are left in the waiting room.")
            return

        current_node = self.head.next
        
        while current_node != self.tail:
            
            print(current_node.elem.name)
            current_node = current_node.next
            
    def reverseTheLine(self):
        
        if self.head.next==self.tail and self.tail.prev==self.head:
            
            print("No patients are left in the waiting room.")
            return

        
        current_node = self.tail.prev
        
        while current_node!=self.head:
            
            print(current_node.elem.name)
            current_node = current_node.prev
        
            
            
def main():
    wrm = WRM()

    while True:
        print("\n1. Add Patient")
        print("2. Serve Patient")
        print("3. Show All Patients")
        print("4. Can Doctor Go Home?")
        print("5. Cancel All Appointments")
        print("6. Reverse the line")

        option=int(input("Enter your choice: "))

        if option==1:
            idt = int(input("Enter patient ID: "))
            name = input("Enter patient name: ")
            age = int(input("Enter patient age: "))
            bloodgroup = input("Enter patient blood group: ")
            wrm.register_patient(idt, name, age, bloodgroup)
            print("Patient is registered successfully.")

        elif option==2:
            wrm.serve_patient()

        elif option==3:
            wrm.show_all_patients()

        elif option==4:
            wrm.can_doctor_go_home()

        elif option==5:
            wrm.cancel_all_appointments()
            
        elif option==6:
            wrm.reverseTheLine()

        else:
            print("Invalid. Please enter again.")


if __name__ == '__main__':
    main()
```
