```python
class Node:
    def __init__(self, elem, next = None):
        self.elem, self.next = elem, next

def create_linked_list(arr):
    head = Node(arr[0])
    tail = head
    for i in arr[1:]:
        new_node = Node(i)
        tail.next = new_node
        tail = new_node
    return head

def count(head):
    count = 0
    while head != None:
        count += 1
        head = head.next
    return count

def print_linked_list(head):
    while head != None:
        print(head.elem, end = '-> ')
        head = head.next
    print('None')
    print()
```

```python
#Do by yourself
def nerdy_run(path,k):

    if len(path)==0 or len(path)==1:
        return None
    
    di={}

    for i in range(len(path)):
        if path[i] in di:
            if i-di[path[i]]<=k:
                return path[i]
        di[path[i]]=i

    return None
```

```python
class Node_pair:
    def __init__(self, key, value, next = None):
        self.key, self.value, self.next = key, value, next


class Hashtable:
    def __init__(self, size):
        self.ht = [None]*size


    def insert(self, s):
        if self.search_hashtable(s) == 'Found':
            print(s,'Already Inserted. Cannot reinsert.')
            print('===============================')
            return
        
        hashed_index = self.__hash_function(s[0])
        pair = Node_pair(s[0], s[1])
        
        if self.ht[hashed_index] == None:
            self.ht[hashed_index] = pair
            
        else:
            pair.next = self.ht[hashed_index]
            self.ht[hashed_index] = pair


    def create_from_array(self, arr):
        for i in arr:
            self.insert(i)

    def print_hashtable(self):
        idx = 0
        for i in self.ht:
            
            print(idx, ':', end = ' ')
            head = i
            
            while head != None:
                print(f'(key: {head.key}, value: {head.value})', end = '-->')
                head = head.next
            print('None')
            print()
            idx += 1


    #Do by yourself
    
    def __hash_function(self, s):
        
        if len(s)%2!=0:
            s+="N"
        i=0 
        j=1
        add=0
        
        for x in range(len(s)//2):
            add+=int(str(ord(s[i])) + str(ord(s[j])))
            i+=2
            j+=2
            
        temp=add%len(self.ht)
    
        return temp
    

    #Do by yourself
    
    def search_hashtable(self, s):
        hash_index=self.__hash_function(s[0])

        if self.ht[hash_index] is None:
            return "Not Found"
        else:
            h = self.ht[hash_index]
            while h is not None:
                if h.key == s[0] and h.value==s[1]:
                    return "Found"
                h = h.next
            return "Not Found"
    
```

```python
class Layered_Hashtable:
    
    def __init__(self, express_array_size):
        self.express_array = [None] * express_array_size

    def print_express_lane(self):
        for i in self.express_array:
            print(i.elem, end = ' '*10)
        print()

    def print_layered_hashtable(self):
        print('Express Lane is:')
        self.print_express_lane()

        for i in range (len(self.express_array)-1):
            node = self.express_array[i]
            next_node = self.express_array[i+1]
            print(f'Normal Lane Nodes between Express Lane Node {node.elem} and Express Lane Node {next_node.elem} are: ')
            while node != next_node:
                print(node.elem, end = '->')
                node = node.next
            print()

        print(f'Normal Lane Nodes ending in the Express Lane Node: {node.elem}')

    #DO IT YOURSELF
    
    def create_layered_hashtable(self, linked_list_head):
        
        bucket=(count(linked_list_head)// len(self.express_array)) + 1
        iter_head=linked_list_head
        i,j=0,0
       
        while iter_head!= None:
            
            if j%bucket == 0:
                self.express_array[i]=iter_head
                i=i+1
                
            iter_head=iter_head.next
            j=j+1
            
        return self.express_array
        

    #DO IT YOURSELF
    
    def search_element(self,k):
        
        bucket=0
        next_bucket_node=0
        
        for i in range(len(self.express_array)-1):
            if self.express_array[i].elem <= k <= self.express_array[i+1].elem:
                bucket=self.express_array[i]
                next_bucket_node=self.express_array[i+1]
                break
    
        
        
        if bucket==0 or next_bucket_node==0:
            
            ola=self.express_array[len(self.express_array)-1].elem
            temp=self.express_array[len(self.express_array)-1]
            
            if ola<k:
    
                while temp!=None:
                    if temp.elem == k:
                        return "Found"
                    temp=temp.next
            else:
                
                return "Not Found"
        
        elif k==bucket.elem or k==next_bucket_node.elem:   
            return "Found"
        
        else:
            while bucket!=next_bucket_node:
                if bucket.elem == k:
                    return "Found"
                bucket=bucket.next
                
        return "Not Found"
        
```

```python
class vehicleNodes:
    def __init__(self, brand, vehicle_type, rent, passenger, next = None):
        self.brand = brand
        self.vehicle_type = vehicle_type
        self.rent = rent
        self.passenger = passenger
        self.next = next
```

```python
# Hashtable using linked list, forward chaining and recurtion 

class VehicleHashTable:
    
    def __init__(self, size):
        self.vehicleTable = [None]*size
        self.empty_slot = {}
        for i in range(size):
            self.empty_slot[i] = False

    def create_from_vehicle_info_array(self, arr):
        for i in arr:
            self.insert_vehicle(i)

    def print_vehicle_hashtable(self):
        idx = 0
        for i in self.vehicleTable:
            print(idx, ':', end = ' ')
            head = i
            while head != None:
                print(f'(Brand: {head.brand}, Type: {head.vehicle_type}, Rent: {head.rent}, No. of Passengers: {head.passenger})', end = '---->')
                head = head.next
            print('None')
            print()
            idx += 1

    def find_empty_slot(self):
        for k,v in self.empty_slot.items():
            idx = k
            del self.empty_slot[k]
            return idx


    #Do it yourself
  
    def __hash_function(self, brand):
        
        index = self.helper(brand)
        
        index %= len(self.vehicleTable)  # Modulo by the size of the hash table
        
        # Do not discard the next 3 lines
        
        if index in self.empty_slot:
            del self.empty_slot[index]
            
        return index 
    
    
    def helper(self,brand,index=0):
        
        if index == len(brand):
            return 0
    
        return ord(brand[index]) + self.helper(brand,index+1)  # Used recursion to find the index


    def insert_vehicle(self, vehicle):
        
        brand, vehicle_type, rent, passenger = vehicle
        vehicle_node = vehicleNodes(brand, vehicle_type, rent, passenger)
        index = self.__hash_function(brand)

        if self.vehicleTable[index] is None:
            
            # If the index is empty, insert the vehicle node at that index
            self.vehicleTable[index] = vehicle_node
            
        else:

            current = self.vehicleTable[index]

            if current.brand == brand:
                
                temp=self.vehicleTable[index]
                vehicle_node.next = temp  # Point the new node's next to the current head's next
                self.vehicleTable[index]=vehicle_node

            # If brand names are not the same, find the next empty index
            
            else:
                temp=self.find_empty_slot()
                self.vehicleTable[temp] = vehicle_node

```

```python
# Hashtable using linear probing and recurtion

def to_binary_recursive(n):
   
    if n == 0:
        return '0'
    elif n == 1:
        return '1'
    else:
        return to_binary_recursive(n // 2) + to_binary_recursive(n % 2)
    

def sum_of_digits_recursive(binary_str, index=0):
    
    if index == len(binary_str):
        return 0
    else:
        return int(binary_str[index]) + sum_of_digits_recursive(binary_str, index + 1)
    

def linear_probing(hash_map, key, value, start_index=0):
    
    
    index = (start_index + key) % len(hash_map)
    
    if hash_map[index] is None:  
        hash_map[index] = value
        
    else:                        
        return linear_probing(hash_map, key, value, start_index + 1)
    
    return hash_map



#=============================Driver Code===============================

HASH_MAP_SIZE = 10
hash_map = [None] * HASH_MAP_SIZE  


input_array = [7, 5, 3, 2]


binary_numbers = [None] * len(input_array)
sums_of_digits = [None] * len(input_array)

for i in range(len(input_array)):
    binary_numbers[i] = to_binary_recursive(input_array[i])
    sums_of_digits[i] = sum_of_digits_recursive(binary_numbers[i])


for i in range(len(sums_of_digits)):
    sum_of_digits = sums_of_digits[i]
    hash_map = linear_probing(hash_map, sum_of_digits, input_array[i])
                                            # Key         # Value


for j in hash_map:
    if j!=None:
        print(j, end=' ')
        
        
```

```python
# Check if there is any duplicate in a tree using hashtable

class BTNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class HashTable:
    def __init__(self, size=101):  # Default size of the hash table is 101
        self.size = size
        self.table = [None] * self.size

    def hash_function(self, key):
        return key % self.size

    def insert(self, key):
        # Using simple linear probing for collision resolution
        hash_index = self.hash_function(key)
        while self.table[hash_index] is not None:
            if self.table[hash_index] == key:  # Duplicate found
                return True
            hash_index = (hash_index + 1) % self.size  # Linear probing
        self.table[hash_index] = key
        return False

    def search(self, key):
        hash_index = self.hash_function(key)
        while self.table[hash_index] is not None:
            if self.table[hash_index] == key:
                return True
            hash_index = (hash_index + 1) % self.size  # Linear probing
        return False

def is_duplicate(root, hash_table):
    if root is None:
        return False
    if hash_table.insert(root.value):
        return True  # Duplicate found
    # Recur for left and right subtrees
    return is_duplicate(root.left, hash_table) or is_duplicate(root.right, hash_table)

def print_ascending_bt(root):
    if root is None:
        return
    # In-order traversal
    print_ascending_bt(root.left)
    print(root.value, end=' ')
    print_ascending_bt(root.right)

# Construct the binary tree from the problem's example
root1 = BTNode(20)
root1.left = BTNode(15)
root1.left.left = BTNode(5)
root1.left.right = BTNode(18)
root1.left.left.right = BTNode(10)
root1.left.right.left = BTNode(22)
root1.right = BTNode(30)
root1.right.left = BTNode(25)
root1.right.right = BTNode(40)
root1.right.left.left = BTNode(22)  # Duplicate node for testing
root1.right.right.left = BTNode(30)  # Duplicate node for testing
root1.right.right.right = BTNode(50)

# Check if there is any duplicate in the binary tree
hash_table = HashTable()
if is_duplicate(root1, hash_table):
    print("True")
else:
    print("False")
    print_ascending_bt(root1)
```
