```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
    row,col = m.shape
    for i in range(row):
        c = 1
        print('|', end='')
        for j in range(col):
            c += 1
            if(len(str(m[i][j])) == 1):
                print(' ',m[i][j], end = '  |')
                c += 6
            else:
                print(' ',m[i][j], end = ' |')
                c += 6
        print()
        print('-'*(c-col))
```

#Zigzag Walk

```python
def walk_zigzag(floor):
    
    r,c=floor.shape
    
    if r%2!=0:
        
        for i in range(c):
            for j in range(r):
                if i%2==0 and j%2==0:
                    print(floor[j][i], end=' ')

            for k in range(r-1,0,-1):
                if i%2!=0 and k%2!=0:
                    print(floor[k][i], end=' ')

            print()
    
    elif r%2==0:
        
        for i in range(c):
            for j in range(r):
                if i%2==0 and j%2==0:
                    #pass
                    print(floor[j][i], end=' ')

            for k in range(r-1,0,-1):
                if i%2!=0 and k%2!=0:
                    #print(k)
                    print(floor[k][i], end=' ')

            print()
        
    pass



floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)

#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6

print('################')
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

#Wall Up Trost District

```python
def create_fence(district, depth):
    district_row, district_col = district.shape
    
    # finding image size after padding
    
    n,m= (district_row+(2*depth)),(district_col+(2*depth))
    
    result=np.zeros((n,m),dtype=int)
    
    r,c=result.shape
    
    
    for i in range(depth,(district_row+depth)):
        for j in range(depth,(district_col+depth)):
            result[i][j]=district[i-depth][j-depth]
            
    for i in range(r):
        for j in range(c):
            if i<depth or i>=n-depth or j<depth or j>=m-depth:
                result[i][j]=8

    
    return result

depth = 1
district = np.array([[2,3,4], [3,4,6], [2,1,4]])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)

#This will print
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
# |  8  |  2  |  3  |  4  |  8  |
# -------------------------------
# |  8  |  3  |  4  |  6  |  8  |
# -------------------------------
# |  8  |  2  |  1  |  4  |  8  |
# -------------------------------
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------

print('################')
depth = 2
district = np.array([
                 [2,3,4,1],
                 [3,4,6,5],
                 [2,1,4,7]
                ])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
```

#Crows vs Cats

```python
import numpy as np

def strength_difference(clubroom):
    r, c = clubroom.shape
    
    size=int(((r*c)-r)/2)  # Length of strength_diff arary without using any loop (Bonus)
    
    strength_diff = np.zeros(size, dtype=int)
    
    count = 0
    
    for i in range(r):
        for j in range(i+1,r):
            strength_diff[count]=clubroom[i][j]-clubroom[j][i]
            count += 1

    return strength_diff

clubroom = np.array([
    [1,  2,  9,  7],
    [4,  5,  1,  8],
    [3,  6,  2,  7],
    [2,  8,  6,  3]
])

print_matrix(clubroom)
returned_value = strength_difference(clubroom)
print('Strength Difference Array is :', returned_value)
unittest.output_test(returned_value, np.array([-2, 6, 5, -5, 0, 1]))
```

#ATM's Triangle

```python
def atm_triangle(n):
    
    atm = np.zeros((n,n),dtype=int)
    
    for i in range(0,n):
        for j in range(0,i+1):
            atm[i,j] = i+1
            if i>=2 and j>0 and j<i:
                sum=0
                for k in range(0,j+1):
                    sum+=atm[i-1,k]
                atm[i,j] = sum

    return atm

def print_without_none(matrix):
    
    row,col = matrix.shape
    
    for i in range(0,row):
        for j in range(0,i+1):
            print(matrix[i,j],end=" ")
        print()

n=6

returned_value = atm_triangle(n)
print_without_none(returned_value)
#This should print
# 1
# 2  2
# 3  4  3
# 4  7  10  4
# 5  11 21  25  5
```

```python
import numpy as np

def points(temp):
    r,c=temp.shape
    count=0
    for i in range(r):
        for j in range(c):
            if temp[i][j]>2:
                if 0<=i-1<r:
                    if temp[i-1][j]==2:
                        count+=1
                if 0<=j-1<c:
                    if temp[i][j-1]==2:
                        count+=1
                if 0<=i+1<r:
                    if temp[i+1][j]==2:
                        count+=1
                if 0<=j+1<c:
                    if temp[i][j+1]==2:
                        count+=1 
    if (count*2)>10:
        print('team won')
    else:
        print('lost')
    return (count*2)
ola=np.array([[0,2,2,0,2],[1,50,2,1,100],[2,2,2,0,2],[0,200,2,0,0]])
print(points(ola))
```
