```python
# Task-1

class Node:
    def __init__(self, data=None):
        self.data = data
        self.prev,self.next=None,None


class Doublylist:
    def __init__(self,lst):
        self.head=Node('X')
        tail=self.head
        self.head.next=tail
        for k in range(len(lst)):
            ev=None
            temp=None
            new=Node(lst[k])
            mode=self.head.next
            
            while mode!=None:
                if mode==self.head:
                    temp='P'
                if temp=='P':
                    break
                if mode.data==new.data:
                    ev='X'
                mode=mode.next 
            if ev=='X':
                pass
                continue
            
            tail.next=new
            new.prev=tail
            tail=new
            tail.next=self.head
```

```python
# Task-2

class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
        self.prev=None
        
class DoublyList:
    def __init__(self,listt=None):
        self.head=None
        self.tail=None
        
        if listt != None:
            temp=Node('X')
            self.head=temp
            self.tail=temp
            for i in listt:
                n=Node(i)
                self.tail.next=n
                n.prev=self.tail
                self.tail=n
                
            self.tail.next=self.head
            
    def showList(self):
        if self.head is None:
            print("Empty linked list!")
        temp=self.head
        i=0
        while temp!=None:
            if temp.next==self.head:
                i+=1
            if i==1:
                print(temp.data, end="")
                print()
                break
            print(temp.data, end="⇌")
            temp=temp.next
    def nodeAt(self, index):
        if index < 0:
            return None
        n = self.head
        for i in range(0, index):
            n = n.next
        return n

    def countNode(self):
        count = 0
        n = self.head
        head = 0
        while head<1:
            if n.next==self.head:
                head+=1
            n = n.next
            count = count + 1
        return count
    
    def insert_index_given(self,newElement,index):
        new_node=Node(newElement)
        c = 0
        it = self.head.next
        while True:
            if it == self.head:
                break
            if c+2 == index:
                break
            c += 1
            it = it.next
            
        if it== self.head:
            print("Not a valid Index")
            return
        
        temp ,it.next= it.next,new_node
        new_node.prev,new_node.next = it,temp
        temp.prev = new_node
        
    def insert(self,newElement):
        new_Node = Node(newElement)
        if self.head is None:
            self.head = new_Node
        else:
            m = self.head.next
            # checking if newElement is already in the linked list or not

            while True:
                if m == self.head:
                    break

                if m.data == new_Node.data:
                    print("Data is already in the linked list!")
                    return
                m = m.next
            # -----------------------------------------------------------
            n = self.head
            while n.next!=self.head:
                n = n.next
            n.next = new_Node
            new_Node.prev = n
            new_Node.next=self.head
            self.head.prev=new_Node

    def remove(self,deleteKey):
        size = self.countNode()
        if size == 1:
            print("Linked List is empty.")
            return
        elif deleteKey < 1 or deleteKey >= size:
            print("Invalid index.Enter a value from 1 to ", size - 1)
            return
        else:
            removeNode = self.nodeAt(deleteKey)
            removeNode.prev.next = removeNode.next
            removeNode.next.prev = removeNode.prev

        removeNode.data = None
        removeNode.next = None
        removeNode.prev = None
        
        
    def removeKey(self,deletekey):
        size = self.countNode()
        if size == 1:
            print("Linked list is empty!")
            return
        m = self.head.next
        while True:
            if m == self.head:
                break

            if m.data == deletekey:
                m.prev.next = m.next
                m.next.prev = m.prev
                print("The deleted key value is: ",m.data)
                m.data = None
                m.next = None
                m.prev = None

                return
            m = m.next

#-----------------Tester------------------
lisst=DoublyList([10,20,30,40,50])
lisst_1=DoublyList([10,20,30,40,50])
lisst_2=DoublyList([10,20,30,40,50])
lisst_3=DoublyList([10,20,30,40,50])
lisst_4=DoublyList([1,2,3,4,5])

# Task (2.2)
lisst.showList()

# Task (2.3)
lisst_2.insert(99)
lisst_2.showList()

# Task (2.4)
lisst_1.insert_index_given(999,2)
lisst_1.showList()

# Task (2.5)
lisst_3.remove(2)
lisst_3.showList()

# Task (2.6)
lisst_4.removeKey(3)
lisst_4.showList()
```
