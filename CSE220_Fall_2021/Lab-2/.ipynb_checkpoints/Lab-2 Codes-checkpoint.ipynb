```python
#Task-1

class Node:
    def __init__(self,data):
        self.data=data        # variable used to stores the data/value/element
        self.next=None        # variable used to reference to link to the next node
class MyList:
    def __init__(self):
        self.head=None        # it is a global variable
            
    def PrintList(self):
        temp=self.head      
        while temp!=None:
            print(temp.data)
            temp=temp.next 
            
#========================Tester Code=========================
            
n1=Node(10)                # Node class e data pathay
n2=Node(20)
n3=Node(30)
n4=Node(40)
n5=Node(50)     

# Calling the list

lisst=MyList()          # print list er 'temp' variable ta ashole Node ke access korte partese ei line tar karone
lisst.head=n1           # head change hoy na just next er value gula change hoy.

# Providing the next Address / Node

n1.next=n2              # eije ekhane Node class er next er value gula change hoche.
n2.next=n3              # next address ba head ki hobe eta diye mention kora hoyeche
n3.next=n4
n4.next=n5

# Printing the list

lisst.PrintList()
```

```python
# Task-2

class Node:
    def __init__(self,element):
        self.data=element
        self.next=None

class MyList:
    def __init__(self,lisst=None):
    
        self.head=None
        self.tail=None

        if lisst!=None:
            for i in lisst:
                temp=Node(i)
                
                if self.tail==None:
                    self.head=temp
                    self.tail=temp
                elif self.tail!=None:
                    self.tail.next=temp        
                    self.tail=temp                                 

    def showList(self):
        ev=MyList()                            
        address=self.head
    
        
        while address!=None:
            
            temp=Node(address.data)
            if ev.head!=None:
                ev.tail.next=temp
                ev.tail=temp
                
            elif ev.head==None:
                ev.head=temp
                ev.tail=temp
            address=address.next
            
        while ev.head!=None:
            if ev.head.next==None:
                print(ev.head.data)
            elif ev.head.next!=None:
                print(ev.head.data, end="->")
            ev.head= ev.head.next

    def isEmpty(self):
        if  self.head!= None:
            print (False)
        
        elif self.head== None:
            print(True) 
        
        
    def clear(self):
        if self.head!=None:
            self.head=None
        
        elif self.head==None:
            print("This list can't be empty")
        
            
    def insert(self,newElement):
    
        current=Node(newElement)
        
        if self.head!=None:
            
            temp1=self.head
            while temp1!=None:
                if temp1.data==current.data:
                    print ("The key already exists. Sorry, can't insert.")
        
                    return temp1.data
                temp1=temp1.next
            
            temp2=self.head
            while temp2.next!=None: 
                temp2=temp2.next             # ekahne basically last er index e amr value ta dhuktese
            temp2.next=current
        
    def insert_index_given(self,newElement,index):          
        if self.head is None:
            print("Linked list cannot be empty")
            return
        size = self.countNode()
        if index < 0 or index >= size:       # size of the linked list is not provided so need to find it
            print("Invalid Index.Enter an index from 0 to", (size - 1))
            return

        newNode = Node(newElement)
        m = self.head
        while m is not None:
            if m.data == newNode.data:
                print("Data is already in the linked list!")
                return
            m = m.next
        newNode = Node(newElement)
        pred = self.nodeAt(index - 1)
        newNode.next= pred.next
        pred.next= newNode
                
    def nodeAt(self, index):
        if index < 0:
            return None
        n = self.head
        for i in range(0, index):
            n = n.next
        return n

    def countNode(self):
        count = 0
        n = self.head
        while n is not None:
            count = count + 1
            n = n.next
        return count
                                       
    def remove(self,deleteKey): 
        size = self.countNode()
        if self.head is None:
            print("Linked List is empty.")
            return
        elif deleteKey < 0 or deleteKey >= size:  # check for index > size later on
            print("Invalid index.Enter a value from 0 to ", size - 1)
            return
        elif deleteKey == 0:
            removedNode = self.head
            self.head = self.head.next

        else:
            pred = self.nodeAt(deleteKey - 1)
            removedNode = pred.next
            pred.next = removedNode.next
        print("The deleted value is :", removedNode.data)
        removedNode.data = None
        removedNode.next = None
        
       
        
        
#========================Tester Code=========================

lisst=MyList([5,6,7,8,9])
lisst_1=MyList([10,20,30,40,50])
lisst_2=MyList([0,1,2,3,4])
lisst_3=MyList([100,200,300,400,500])

# Task (2.2)
lisst.showList()

# Task (2.3)
lisst.isEmpty()

# Task (2.4)
print(lisst.clear())

# Task (2.5)
lisst_2.insert(2)
lisst_2.showList()

# Task (2.6)
lisst_1.insert_index_given(877,3)
lisst_1.showList()

# Task (2.7)
lisst_3.remove(2)
lisst_3.showList()
```

```python
# Task-3

class Node:
    def __init__(self,element):
        self.data=element
        self.next=None

class MyList:
    def __init__(self,lisst=None):
    
        self.head=None 
        self.tail=None

        if lisst!=None:
            for i in lisst:
                temp=Node(i)             # ekhanei main khela. Temp ta ekhn pura Node class er access niye nise.
                                         # Temp diye Node class er data use korte parbo.
                if self.tail==None:
                    self.head=temp
                    self.tail=temp       # ei line er madddhome self.tail 'Node' class er access pache karon temp=Node(i)
                    
                elif self.tail!=None:
                    self.tail.next=temp  # suppose self.tail=23 chilo and 2nd element ta 45 tahole list onujai 23,45 hbe     
                    self.tail=temp       # so, ei list er tail hoche 45 and (self.tail=23).next is also 45 karon                     
                                         # Amr 23 er next element is 45. ei jonno both self.tail,self.tail.next= temp(45)
    
    
    def even(self):
        
        ev=MyList()       # agee jmn temp diye Node class access korsi ekhn ev diye Mylist class access korbo
        
        start=self.head
    
        if self.head!=None:
            while start!=None:          # ekahne start tare ami upor niche korbo tai while func e falailam
                if start.data%2==0:     # start.head Node class er data ke access korte partese karon 
                                        # Mylist class e amra self.head=temp dhorechilam and temp=Node(i) chilo
                        
                    temp=Node(start.data) 
                    if ev.tail==None:        # These things are same as Mylists alogorithm
                        ev.head=temp
                        ev.tail=temp
                    elif ev.tail!=None:
                        ev.tail.next=temp    # Basically amra ei line diye data ke Node class e dhukachi ebng 
                        ev.tail=temp         # ev.head.next diye print dichi.
                    
                start=start.next             # next data/element ke access korche. ek shomoy eta none hoye jabe
                                             # Ebong while loop ke break kore dibe 
            ev.PrintList()
        
        #while ev.head!=None:
            
            #if ev.head.next!=None:            # ev.head.next Node class er data ke access korte partese karon 
                                               # Mylist class e amra self.head=temp dhorechilam and temp=Node(i) chilo
                #print(ev.head.data,end="->")
            
            #elif ev.head.next==None:
                #print(ev.head.data)           # shudhu ev.head.data print hbe karon ev.head.next too none hoye jay.
            
            #ev.head= ev.head.next
    
    def reverse(self):
        prev = None
        current = self.head         # suppose amr list hoche [1,2,3,4,5]
        while current!=None:
            temp= current.next      # next digit take ei variable e store korlam. For, example '2' 
            current.next = prev     # ekhance basically 2 er jaygay 1 ke neya hoche.
            prev = current          # previous ba ager ghor take ami ekhn new head announce korlam
            current = temp          # temp e jei '2' store chilo shetake ekhn annouce kora new head e boshalam.
        self.head=prev  
        
        
    def in_or_out_list(self,num):
        temp=self.head
        
        while temp!=None:
            if temp.data==num:
                 return True       # always return er khetre je kono ekta return use hbe
            elif temp.data!=num:   # if condition fullfill korle true return hbe r naile always false return hbe
                temp=temp.next
        return False   
    
    
    def sortlist(self):
        num_1= self.head
        while num_1.next!=None:       # ei while loop diye num_1 control kora hobe
            
            num_2= num_1.next         # 2nd element ke store korlam variable e

            while num_2!=None:        # ei while loop diye num_2 control kora hobe 
                if num_1.data>num_2.data:
                    temp=num_1.data
                    num_1.data=num_2.data     # simply interchange kora hoche ba agee piche neya holo data.
                    num_2.data=temp
                num_2=num_2.next
            num_1=num_1.next          
            
    def rotate(self,rt,k):
        if rt=='left':
            if k == 0: 
                return
            current = self.head                  # first loop run hoyar por amr current=3 holo
            cnt = 1
            while(cnt<k and current!=None):      # ei loop diye basically rotate korar por jei 
                current = current.next           # head pabo sheta findout kora hoche.
                cnt += 1
                
            temp=current                         # pore shei head ke ekta variable e store kora hobe.
            
            while current.next!=None:            # ekhane 3 er por ja ase ogula print holo.....jmn 4,5 etc
                current=current.next              
            
             
            current.next = self.head             # amr last current 
            self.head = temp.next
            temp.next = None

            
            
        elif rt=='right':
            if k == 0: 
                return
            current = self.head
            cnt = 1
            while current.next:
                current = current.next
                cnt += 1
            current.next= self.head
            prelen = (cnt-(k%cnt))-1
            pre = self.head
            #if current is None:
                #return

            while prelen>0:
                pre = pre.next
                prelen-=1
            self.head = pre.next
            pre.next=None 
    
    def addition(self):
        add=0
        temp=self.head
        
        while temp!=None:
            add=add+(int(temp.data))
            temp=temp.next
        print('Sum:',add)    
        
    def PrintList(self):
        temp=self.head  
        while temp!=None:       
            if temp.next!=None:
                print(temp.data, end="-> ")
            
            elif temp.next==None:                 
                print(temp.data)
            
            temp=temp.next     
            
    
        
#========================Tester Code=========================


lisst=MyList([1,2,3,4,5])

# Task (3.1)
lisst.even()

# Task (3.3)
lisst.reverse()
lisst.PrintList()

# Task (3.4)
lisst.sortlist()
lisst.PrintList()

# Task (3.5)
lisst.addition()

# Task (3.6)        
l1=MyList([1,2,3,4,5])
l1.rotate('right',2)
l1.PrintList()           
l2=MyList([1,2,3,4,5])
l2.rotate('left',2)
l2.PrintList()

# Task (3.2)
lisst.in_or_out_list(2)
```
